```python
import os 
import cv2#opencv
import glob
from tqdm.notebook import tqdm 
import numpy as np

nomes_arquivos_NORMAL = glob.glob(os.path.join('train/NORMAL', '*')) + glob.glob(os.path.join('test/NORMAL', '*'))#os.listdir('test/NORMAL')
nomes_arquivos_PNEUMONIA =  glob.glob(os.path.join('train/PNEUMONIA', '*')) + glob.glob(os.path.join('test/PNEUMONIA', '*'))#os.listdir('train/PNEUMONIA') + os.listdir('test/PNEUMONIA')

x_data = []
y_data = []
for nomal in tqdm(nomes_arquivos_NORMAL):
    img_normal = cv2.imread(nomal, cv2.IMREAD_GRAYSCALE)
    new_width = 128
    new_height = 128
    resized_image = cv2.resize(img_normal, (new_width, new_height))
    resized_image = np.array(resized_image/255)
    x_data.append(resized_image)
    #y_data.append('normal')
    y_data.append(0)

for pneumonia in tqdm(nomes_arquivos_PNEUMONIA):
    img_pneumonia = cv2.imread(pneumonia, cv2.IMREAD_GRAYSCALE)
    new_width = 128
    new_height = 128
    resized_image = cv2.resize(img_pneumonia, (new_width, new_height))
    resized_image = np.array(resized_image/255)
    x_data.append(resized_image)
    y_data.append(1)
x_data = np.array(x_data)
y_data = np.array(y_data)

from sklearn.model_selection import train_test_split

# Divisão em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=42)


import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models  import Sequential

model = Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
# Compilar o modelo
model.compile(optimizer='adam',
loss='binary_crossentropy',
metrics=['accuracy'])

# Treinar o modelo
model.fit(np.array(X_train), np.array(y_train), epochs=10, validation_data=(X_test, y_test))

import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Obter as previsões do modelo
y_pred = model.predict(X_test)  # X_test são seus dados de teste

# Converter as previsões para classes binárias (0 ou 1)
y_pred_classes = np.round(y_pred).astype(int)

# Converter os rótulos de teste para classes binárias
y_test_classes = y_test.astype(int)  # y_test são seus rótulos de teste

# Calcular as métricas
accuracy = accuracy_score(y_test_classes, y_pred_classes)
precision = precision_score(y_test_classes, y_pred_classes)
recall = recall_score(y_test_classes, y_pred_classes)
f1 = f1_score(y_test_classes, y_pred_classes)

# Calcular a matriz de confusão
confusion_matrix = confusion_matrix(y_test_classes, y_pred_classes)

# Imprimir as métricas e a matriz de confusão
print("Acurácia: {:.2f}".format(accuracy))
print("Precisão: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1-score: {:.2f}".format(f1))
print("Matriz de Confusão:")
print(confusion_matrix)


```
